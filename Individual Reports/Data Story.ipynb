# Data Story

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
wine = pd.read_csv('wine_clean.csv')
```

## What are some descriptive statistics of the data?

```python
print(wine.info())
```

```python
print(wine.describe())
```

While the mean and median values of the 'points' column are very similar, at 88.42 and 88 points respectively,  the mean and median values of the 'price' column seem to show a a larger difference, at 35.14 and 25 respectively, which could mean that there are outliers skewing the data.

Additionally, the min and max values for the price column show a greater spread of values than that of the points column. The high max in the price column also points to high value outliers.

## Histograms

```python
plt.hist(wine['points'])
plt.title('Points Distribution')
plt.show()

```

The points column seems to have normally distributed data

```python
plt.hist(wine['price'])
plt.title('Price Distribution')
plt.show()
```

The Price column is not normally distributed and might need to be transformed via logarithm 

## Scatter Plot

```python
sns.relplot(x="price", y="points", data=wine)
plt.show()
```

While there seems to be a positive correlation between price and points, it does not seem to be a strong correlation

## Bar Graph

```python
price_range = []


for row in wine['price']:
    if row <=29:
        price_range.append('0-29'),
    elif (row <=59) & (row > 29):
        price_range.append('30-59'),
    elif (row<=99) & (row > 59):
        price_range.append('60-99'),
    elif (row<=499) & (row > 99):
        price_range.append('100-499'),
    else:
        price_range.append('500-1,000')

wine['price_range'] = np.array(price_range)



points_byprice = wine.groupby('price_range').mean()

reordered= (['0-29', '30-59' , '60-99', '100-499', '500-1,000'])

points_byprice= points_byprice.reindex(reordered)

print(points_byprice)
sns.barplot(x=points_byprice.index , y='points', data=points_byprice)
plt.xlabel('Wine Price ($)')
plt.ylabel('Wine Review Score Average')
plt.show()
```

As price increases, the average review increases as well.

```python
sns.distplot(wine['price'], hist = False)
plt.xlabel('Wine Price ($)')
plt.show()
```

## Pearson correlation coefficient

```python
np.corrcoef(x =wine['price'] , y=wine['points'])[0,1]
```

The correlation between price and points is moderate and positive

## Hypotheses

A predictor for how well a bottle of wine will perform, is its price. As the price goes up, so does the review score. However this relationship is not that strong by itself, there needs to be further variables that need to be used to find a stronger correllation.

A reason as to why price does not correlate strongly with the review scores might have to do with high priced outliers not representing a proportional review score as those wines who are either low or modestly priced.

In the future, the price data might have to be normalized with a logarithm in order to properly use it in a linear regression model since it is skewed to the right according to the histogram and distribution plots.
